**Question 1:**
**Find out the maximum likelihood parameters values for california_housing [‘latiyude’,total_rooms’,’house_median_age’,’total_bedrooms’] data, each column individually.**

```python
import pandas as pd
import numpy as np

data = pd.read_csv('/content/housing.csv')

columns = ['latitude', 'total_rooms', 'housing_median_age', 'total_bedrooms']

for column in columns:
    feature = data[column].dropna()
    mean_mle = np.mean(feature)
    variance_mle = np.var(feature)

    print(f"Maximum Likelihood Estimates for {column}:")
    print(f"Mean (μ): {mean_mle:.2f}")
    print(f"Variance (σ²): {variance_mle:.2f}")
    print()
```

**Question 2:**
**Find log-likelihood values for 50, 75, and 80. For California_housing [‘latitude’] and find the Max likelihood values**

```python
import pandas as pd
import numpy as np

data = pd.read_csv('/content/housing.csv')
latitude = data['latitude'].dropna()

mean_mle = np.mean(latitude)
variance_mle = np.var(latitude)
std_dev_mle = np.sqrt(variance_mle)

log_likelihood_50 = -0.5 * np.log(2 * np.pi * variance_mle) - 0.5 * ((50 - mean_mle) ** 2) / variance_mle
log_likelihood_75 = -0.5 * np.log(2 * np.pi * variance_mle) - 0.5 * ((75 - mean_mle) ** 2) / variance_mle
log_likelihood_80 = -0.5 * np.log(2 * np.pi * variance_mle) - 0.5 * ((80 - mean_mle) ** 2) / variance_mle

print(f"Log-Likelihood at 50: {log_likelihood_50:.2f}")
print(f"Log-Likelihood at 75: {log_likelihood_75:.2f}")
print(f"Log-Likelihood at 80: {log_likelihood_80:.2f}")
print(f"Maximum Likelihood Mean (μ): {mean_mle:.2f}")
print(f"Maximum Likelihood Variance (σ²): {variance_mle:.2f}")
```

**Question 3:**
**Estimate the density of each column of california_housing [‘latitude’,total_rooms’,’house_median_age’,’total_bedrooms’] data, For the bins [5, 10, 15, 20]. And write the observations.**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = '/content/sample_data/california_housing_test.csv'
data = pd.read_csv(file_path)

# List of columns to analyze - Correct the column name here
columns = ['latitude', 'total_rooms', 'housing_median_age', 'total_bedrooms']
# List of bin sizes
bin_sizes = [5, 10, 15, 20]

for column in columns:
    for bins in bin_sizes:
        plt.figure(figsize=(10, 6))
        plt.hist(data[column].dropna(), bins=bins, density=True, alpha=0.7, edgecolor='black')
        plt.title(f"Density of {column} with {bins} Bins")
        plt.xlabel(column)
        plt.ylabel("Density")
        plt.grid(True)
        plt.show()
```
